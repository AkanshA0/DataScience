```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_ban = pd.read_csv(r'D:\A_semester_6\SN phase 2\data\Bangalore.csv')
df_chn = pd.read_csv(r'D:\A_semester_6\SN phase 2\data\Chennai.csv')
df_del = pd.read_csv(r'D:\A_semester_6\SN phase 2\data\Delhi.csv')
df_hyd = pd.read_csv(r'D:\A_semester_6\SN phase 2\data\Hyderabad.csv')
df_kol = pd.read_csv(r'D:\A_semester_6\SN phase 2\data\Kolkata.csv')
df_mum = pd.read_csv(r'D:\A_semester_6\SN phase 2\data\Mumbai.csv')
```

```python
dfb=pd.DataFrame(df_ban.groupby(['No. of Bedrooms'])['No. of Bedrooms','Price'].mean())
dfc=pd.DataFrame(df_chn.groupby(['No. of Bedrooms'])['No. of Bedrooms','Price'].mean())
dfd=pd.DataFrame(df_del.groupby(['No. of Bedrooms'])['No. of Bedrooms','Price'].mean())
dfh=pd.DataFrame(df_hyd.groupby(['No. of Bedrooms'])['No. of Bedrooms','Price'].mean())
dfk=pd.DataFrame(df_kol.groupby(['No. of Bedrooms'])['No. of Bedrooms','Price'].mean())
dfm=pd.DataFrame(df_mum.groupby(['No. of Bedrooms'])['No. of Bedrooms','Price'].mean())
plt.figure(figsize=(15,10)) 
plt.plot(dfb['No. of Bedrooms'],dfb['Price'])
plt.plot(dfc['No. of Bedrooms'],dfc['Price'])
plt.plot(dfd['No. of Bedrooms'],dfd['Price'])
plt.plot(dfh['No. of Bedrooms'],dfh['Price'])
plt.plot(dfk['No. of Bedrooms'],dfk['Price'])
plt.plot(dfm['No. of Bedrooms'],dfm['Price'])
plt.legend(['Bangalore','Chennai','Delhi','Hyderabad','Kolkata','Mumbai'])
plt.xlabel('No. of Bedrooms')
plt.ylabel('Average Price')
plt.title('Number of bedrooms Vs Average price')
plt.grid(True)
plt.show()
```

```python
#df_ban.describe()

hl=df_ban.columns
headlist=hl[4:]

df1=pd.DataFrame(columns=['Price','facilities'],dtype=float)
df2=pd.DataFrame(columns=['Price','facilities'],dtype=float)
df3=pd.DataFrame(columns=['Price','facilities'],dtype=float)
df4=pd.DataFrame(columns=['Price','facilities'],dtype=float)
df5=pd.DataFrame(columns=['Price','facilities'],dtype=float)
df6=pd.DataFrame(columns=['Price','facilities'],dtype=float)

count=0
for i,r in df_ban.iterrows():
    for l in headlist:
        if r[l]>=1:
            count=count+1
    df1.loc[i]=[r['Price'],count]
    #print(i,r['Price'],count)
    count=0
#df1
            
count=0
for i,r in df_chn.iterrows():
    for l in headlist:
        if r[l]>=1:
            count=count+1
    df2.loc[i]=[r['Price'],count]
    #print(i,r['Price'],count)
    count=0
#df2
            
count=0
for i,r in df_del.iterrows():
    for l in headlist:
        if r[l]>=1:
            count=count+1
    df3.loc[i]=[r['Price'],count]
    #print(i,r['Price'],count)
    count=0
#df3

count=0
for i,r in df_hyd.iterrows():
    for l in headlist:
        if r[l]>=1:
            count=count+1
    df4.loc[i]=[r['Price'],count]
    #print(i,r['Price'],count)
    count=0
#df4

count=0
for i,r in df_kol.iterrows():
    for l in headlist:
        if r[l]>=1:
            count=count+1
    df5.loc[i]=[r['Price'],count]
    #print(i,r['Price'],count)
    count=0
#df5

count=0
for i,r in df_mum.iterrows():
    for l in headlist:
        if r[l]>=1:
            count=count+1
    df6.loc[i]=[r['Price'],count]
    #print(i,r['Price'],count)
    count=0
#df6

```

```python
df1_res=pd.DataFrame(df1.groupby(['facilities'])['facilities','Price'].mean())
df2_res=pd.DataFrame(df2.groupby(['facilities'])['facilities','Price'].mean())
df3_res=pd.DataFrame(df3.groupby(['facilities'])['facilities','Price'].mean())
df4_res=pd.DataFrame(df4.groupby(['facilities'])['facilities','Price'].mean())
df5_res=pd.DataFrame(df5.groupby(['facilities'])['facilities','Price'].mean())
df6_res=pd.DataFrame(df6.groupby(['facilities'])['facilities','Price'].mean())

plt.figure(figsize=(15,10)) 
plt.plot(df1_res['facilities'],df1_res['Price'],
    linewidth=1,
    linestyle='dotted',
    color='g',
    marker='>',
    markersize=5,
    markerfacecolor='green')
plt.plot(df2_res['facilities'],df2_res['Price'],
    linewidth=1,
    linestyle='dotted',
    color='r',
    marker='<',
    markersize=5,
    markerfacecolor='red')
plt.plot(df3_res['facilities'],df3_res['Price'],
    linewidth=1,
    linestyle=':',
    color='y',
    marker='*',
    markersize=5,
    markerfacecolor='yellow')
plt.plot(df4_res['facilities'],df4_res['Price'],
    linewidth=1,
    linestyle='-.',
    color='black',
    marker='s',
    markersize=5,
    markerfacecolor='white')
plt.plot(df5_res['facilities'],df5_res['Price'],
    linewidth=1,
    linestyle='--',
    color='violet',
    marker='o',
    markersize=5,
    markerfacecolor='black')
plt.plot(df6_res['facilities'],df6_res['Price'],
    linewidth=1,
    linestyle='--',
    color='b',
    marker='^',
    markersize=5,
    markerfacecolor='pink')

plt.legend(['Bangalore','Chennai','Delhi','Hyderabad','Kolkata','Mumbai'])
plt.xlabel('Variety of facilities')
plt.ylabel('Average Price')
plt.title('Variety of facilities available Vs Average price')
plt.grid(True)
plt.show()
```

```python
max_min = [[df_ban['Price'].max(), df_chn['Price'].max(), df_del['Price'].max(), df_hyd['Price'].max(), df_kol['Price'].max(), df_mum['Price'].max()],
           [df_ban['Price'].min(), df_chn['Price'].min(), df_del['Price'].min(), df_hyd['Price'].min(), df_kol['Price'].min(), df_mum['Price'].min()]]

x_pos = np.arange(6)
fig = plt.figure()

# add_axes([xmin,ymin,dx,dy]) 
ax = fig.add_axes([0,0,1,1])
 
ax.bar(x_pos - 0.1, max_min[0], width=0.3, color='green')
ax.bar(x_pos + 0.1, max_min[1], width=0.3, color='red')

plt.legend(['Max Price','Min Price'])
plt.xlabel('Cities')
plt.ylabel('Price')
plt.title('maximum and minimum price in cities')
plt.grid(True)
plt.show()
```
